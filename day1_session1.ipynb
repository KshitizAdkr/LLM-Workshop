## Sentiment Analysis using Transformers Models

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    ["I've been waiting for whole life.",
     "I hate this so much"
     ]
)
```

## Text generation model (like ChatGPT)

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("Once upon atime in national college of engineering..")
```

## Language Translator

```python
translator =  pipeline("translation_en_to_de")
translator("I love programming language")
```

## NER Models

```python
ner = pipeline("ner", model = "dbmdz/bert-large-cased-finetuned-conll03-english")
ner("I am Kai and I am from Kathmandu Nepal")
```
